```python
import pandas as pd
import statsmodels.formula.api as smf
```

```python
df = pd.read_excel('ceosal1.xls',
                   header=None,
                   usecols=[0, 3],
                   names=['salary', 'roe'])
```

```python
df.head()
```

### Comandos sugeridos

#### Regressão por Mínimos Quadrados Ordinários (MQO ou OLS na sigla em inglês)

```python
modelo = smf.ols(formula='salary ~ roe', data=df)
reg = modelo.fit()
```

```python
tabela = pd.DataFrame({'roe': df['roe'][:15], 
                       'salary': df['salary'][:15],
                       'salchapeu': reg.fittedvalues[:15],
                       'uchapeu': reg.resid[:15]})
```

```python
tabela.index = range(1, 16)
```

```python
# Tabela 2.2 Valores estimados e resíduos dos 15 primeiros CEOs
tabela
```

```python
# CEOs com salários menores do que os previstos (considerando-se as 15 primeiras observações)
tabela[tabela['uchapeu'] < 0]
```

```python
# CEOs com salários maiores do que os previstos (considerando-se as 15 primeiras observações)
tabela[tabela['uchapeu'] >= 0]
```
